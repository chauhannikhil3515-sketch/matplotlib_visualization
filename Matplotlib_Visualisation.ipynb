```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```

```python
plt.plot(x,y, color="red", linestyle="--", marker="o", linewidth=3,markersize=9)  #to understand matplot functions search on google matplotlib.pyplot plot function doand read answer given by AI at top
plt.title("Line Plot")                                                            
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)   #if we want grid we have to use this function
```

```python
plt.figure(figsize=(10,7))      #to avoid overlapping we have given figsize(10,7)
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("First Graph")

plt.subplot(2,2,2)
plt.plot(x,y)
plt.title("Second Graph")

plt.subplot(2,2,3)
plt.plot(x,y,color="orange")
plt.title("Third Graph")

plt.subplot(2,2,4)
plt.plot(x,y)
plt.title("Fourth Graph")
```

```python
plt.figure(figsize=(5,12))      #to avoid overlapping we have given figsize(10,7)
plt.subplot(3,1,1)
plt.plot(x,y)
plt.title("First Graph")

plt.subplot(3,1,2)
plt.plot(x,y)
plt.title("Second Graph")

plt.subplot(3,1,3)
plt.plot(x,y,color="orange")
plt.title("Third Graph")
```

## BAR PLOT

```python
Categories=['A','B','C','D','E']
Values=[5,7,3,8,6]
plt.bar(Categories,Values, color="red")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Graph")
```

```python
#if we want to fill different colours in each bar we have to give colors in the form of list 
Categories=['A','B','C','D','E']
Values=[5,7,3,8,6]
plt.bar(Categories,Values, color=['red','yellow','pink','green','violet'])
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Graph")
```

```python
#if we want this vertical bar graph to be horizontal in syntax write plt.barh()
Categories=['A','B','C','D','E']
Values=[5,7,3,8,6]
plt.barh(Categories,Values, color=['red','yellow','pink','green','violet'])
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Graph")
```

**PIE CHART**

```python
labels=['A','B','C','D']
sizes=[30,20,40,10]
plt.pie(sizes,labels=labels, colors=['gold','green','lightskyblue','yellowgreen'],autopct="%0.2f%%") 
#here autopct used for beautifying and making it  understantable, %=formatting, 0.2=means after decimal we need 2 values after percentage
# f= representing float, %%=we want data in %form 
```

```python
#we can play around this pie chart by using shadow,explode functions;-
labels=['A','B','C','D']
sizes=[30,20,40,10]
explode=[0.2,0,0,0]  #jis place pe value dalenge vo explode hoga jese yha first place pe 0.2 dali,first value equal to 30 in sizes list, so 30 vala part hi explode hoga 
plt.pie(sizes,labels=labels, colors=['gold','green','lightskyblue','yellowgreen'],autopct="%0.2f%%",shadow=True,explode=explode) 
```

**BOX PLOT**- in this outlier will be outside the boxplot 


outlier= the number which is extrmely out of the range of other numbers,this outlier number can disturb the overall avg of numbers

```python
marks=[35,40,42,45,48,50,55,60,65,90]   #here as we can see in graph 90 is outlier, lie outside the range of other numbers
plt.boxplot(marks)                      #so this outlier (90) can disturb the data it will not let you find the clear picture of the data
plt.title('Box Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
```

## Top 3 Countries

```python
df['Country'].value_counts().head(3)
```

```python
#above, left side data (Country) is called index
#right side data(country count) is called values
#we want these index and values in list form so we can make pie chart 

country_name=df['Country'].value_counts().head(3).index
country_name
```

```python
country_count=df['Country'].value_counts().head(3).values
country_count
```

```python
#Now make Pie Chart because values are less than 5 when values are less that 5 always make pie chart otherwise bar chart
plt.pie(country_count,labels=country_name,autopct='%0.2f%%')
```
